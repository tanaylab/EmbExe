#### Generate cell type composition barplots

```r
barplot_cell_type_composition = function(mat_nm,ignore_embryos = NULL) {


    load(sprintf("data/%s/color_annotation/cmp_annot.Rda",mat_nm))
    query_cls_col = cmp_annot$query_cls_col
    
    mat_wt = scdb_mat("embexe")
    mc_wt = scdb_mc("embexe_recolored")
    mat_query = scdb_mat(mat_nm)
    
    df_time = fread(sprintf("data/%s/time_match/time_match_summary.txt",mat_nm))
    
    df_time
    ko_embs = df_time$embryo
    
    if(!is.null(ignore_embryos)) {
        ko_embs = setdiff(ko_embs,ignore_embryos)
    }
    
    query_cls = names(query_cls_col)
    query_cls = query_cls[mat_query@cell_metadata[query_cls,"embryo"] %in% ko_embs]    


    emb_cols <- mc_wt@color_key$color[1:41]

    exe_cols <- mc_wt@color_key$color[43:50]


    embryo_to_col_ko <- table(factor(x = query_cls_col[query_cls],
                                 levels = mc_wt@color_key$color),
                          mat_query@cell_metadata[query_cls,"embryo"])


    embryo_to_col_ko_emb <- embryo_to_col_ko[emb_cols,] # subset of embryonic cell types

    embryo_to_col_ko_exe <- embryo_to_col_ko[exe_cols,] # subset of ExE cell types


    embryo_to_col_ko_emb_n <- t(t(embryo_to_col_ko_emb)/colSums(embryo_to_col_ko_emb))

    embryo_to_col_ko_exe_n <- t(t(embryo_to_col_ko_exe)/colSums(embryo_to_col_ko_exe))

    par(mar = c(16,3,2,2))
    barplot(embryo_to_col_ko_emb_n,
        col = rownames(embryo_to_col_ko_emb_n),
        las = 2)


    barplot(embryo_to_col_ko_exe_n,
        col = rownames(embryo_to_col_ko_exe_n),
        las = 2)
    
}
```

```r
mc_wt@color_key
```

```r
exe_mesoderm_cell_type_color = mc_wt@color_key$color[c(23,24,26,27)]

load("data/Bmp4_KO_embexe_wt/color_annotation/cmp_annot.Rda")
Bmp4_KO_embexe_cell_type_color_annotation = cmp_annot$query_cls_col
```

```r
Bmp4_KO_embexe_exe_mesoderm_cells = names(Bmp4_KO_embexe_cell_type_color_annotation)[Bmp4_KO_embexe_cell_type_color_annotation %in% exe_mesoderm_cell_type_color]
```

```r
table(mat_bmp4_embexe@cell_metadata[ Bmp4_KO_embexe_exe_mesoderm_cells,"embryo"])
table(mat_bmp4_embexe@cell_metadata[ colnames( mat_bmp4_embexe@mat),"embryo"])
setdiff(names(table(mat_bmp4_embexe@cell_metadata[ colnames( mat_bmp4_embexe@mat),"embryo"])),names(table(mat_bmp4_embexe@cell_metadata[ Bmp4_KO_embexe_exe_mesoderm_cells,"embryo"])))
```

```r
bmp4_level_per_embryo = tapply(mat_bmp4_embexe@mat["Bmp4",Bmp4_KO_embexe_exe_mesoderm_cells],
                               mat_bmp4_embexe@cell_metadata[ Bmp4_KO_embexe_exe_mesoderm_cells,"embryo"],sum)/as.vector(
                        tapply(colSums(mat_bmp4_embexe@mat[,Bmp4_KO_embexe_exe_mesoderm_cells]),mat_bmp4_embexe@cell_metadata[ Bmp4_KO_embexe_exe_mesoderm_cells,"embryo"],sum))
```

```r
plot(log2(bmp4_level_per_embryo + 1e-5))
```

```r
germline_ko_filtered_embryos = names(bmp4_level_per_embryo)[log2(bmp4_level_per_embryo + 1e-5) <  -14]
```

```r
barplot_cell_type_composition("Bmp4_KO_embexe_wt",ignore_embryos = setdiff(germline_ko_embryos,germline_ko_filtered_embryos))
```
