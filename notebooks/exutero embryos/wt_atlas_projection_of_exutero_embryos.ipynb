### Finding best matched wildtype embryo time range for exutero embryos
We calculate **metacell** vs **transcriptional_rank** distribution **mc_ag** and smooth it over several neighboring time points.
Exutero metacell distribution is then matched with the best correlated **mc_ag_smoothed** column for all **embryonic_metacells**.

```r
wt_cells_exe = names(mc_wt_exe@mc)[!is.na(mat_wt@cell_metadata[names(mc_wt_exe@mc),"transcriptional_rank"])]
wt_cells_emb = names(mc_wt_emb@mc)[!is.na(mat_wt@cell_metadata[names(mc_wt_emb@mc),"transcriptional_rank"])]
mc_ag = table(mc_wt_emb@mc[wt_cells_emb],mat_wt@cell_metadata[wt_cells_emb,"transcriptional_rank"])

mc_ag_smoothed = t(apply(mc_ag,1,function(v) {
    return(rollmean(v,k = 41))
}))

mc_ag_smoothed_11 = t(apply(mc_ag,1,function(v) {
    return(rollmean(v,k = 11))
}))
```

```r
exut_emb_cor = tgstat::tgs_cor(cbind(hf_exut_freq,ls_exut_freq),mc_ag_smoothed)
exut_emb_cor_11 = tgstat::tgs_cor(cbind(hf_exut_freq,ls_exut_freq),mc_ag_smoothed_11)
```

```r
emb_dist = as.matrix(dist(rbind(hf_exut_freq,ls_exut_freq,t(mc_ag_smoothed))))
emb_dist_11 = as.matrix(dist(rbind(hf_exut_freq,ls_exut_freq,t(mc_ag_smoothed_11))))
dim(emb_dist)
dim(emb_dist_11)

plot(emb_dist[2,3:213])
plot(emb_dist_11[1,3:243])
```

```r
best_matching_late_streak_rank = as.numeric(names(which.min(emb_dist[2,3:ncol(emb_dist)])))
message("Best matching WT transcriptional of late streak ex utero embryos: ",best_matching_late_streak_rank)
best_matching_head_fold_rank = as.numeric(names(which.min(emb_dist[1,3:ncol(emb_dist)])))
message("Best matching WT transcriptional of late head fold ex utero embryos: ",best_matching_head_fold_rank)
```

```r
#lhf_embryos = intersect(exe_embryos,morphology_embryos$embryo[morphology_embryos$morphological_stage == "late head fold"])
wt_hf_cells_emb = wt_cells_emb[(mat_wt@cell_metadata[wt_cells_emb,"transcriptional_rank"] %in% c((best_matching_head_fold_rank - 3):(best_matching_head_fold_rank + 3))) & (mat_wt@cell_metadata[wt_cells_emb,"Experiment"] != "Wildtype project")]
wt_hf_cells_exe = wt_cells_exe[(mat_wt@cell_metadata[wt_cells_exe,"transcriptional_rank"] %in% c((best_matching_head_fold_rank - 3):(best_matching_head_fold_rank + 3))) & (mat_wt@cell_metadata[wt_cells_exe,"Experiment"] != "Wildtype project")]


#wt_ls_cells = wt_cells[(ref_mat@cell_metadata[wt_cells,"transcriptional_rank"] %in% c(2)) & (ref_mat@cell_metadata[wt_cells,"Experiment"] != "Wildtype project")]
wt_ls_cells_emb = wt_cells_emb[(mat_wt@cell_metadata[wt_cells_emb,"transcriptional_rank"] %in% c((best_matching_late_streak_rank - 15):(best_matching_late_streak_rank + 15))) & (mat_wt@cell_metadata[wt_cells_emb,"Experiment"] != "Wildtype project")]
wt_ls_cells_exe = wt_cells_exe[(mat_wt@cell_metadata[wt_cells_exe,"transcriptional_rank"] %in% c((best_matching_late_streak_rank - 15):(best_matching_late_streak_rank + 15))) & (mat_wt@cell_metadata[wt_cells_exe,"Experiment"] != "Wildtype project")]


length(wt_hf_cells_emb)
length(wt_ls_cells_emb)
length(wt_hf_cells_exe)
length(wt_ls_cells_exe)

matched_wt_hf_cells = c(wt_hf_cells_emb,wt_hf_cells_exe)
matched_wt_ls_cells = c(wt_ls_cells_emb,wt_ls_cells_exe)
matched_wt_cells_emb = c(wt_hf_cells_emb,wt_ls_cells_emb)
matched_wt_cells_exe = c(wt_hf_cells_exe,wt_ls_cells_exe)
```

```r
df_exutero = data.frame(cell = c(matched_wt_cells_emb,matched_wt_cells_exe),
                        exutero_batch = c(rep("HF",length(wt_hf_cells_emb)),rep("LS",length(wt_ls_cells_emb)),
                                          rep("HF",length(wt_hf_cells_exe)),rep("LS",length(wt_ls_cells_exe))))

write.table(x = df_exutero,file = "data/exutero_matched_wt_cells.tsv",sep = '\t')
```
