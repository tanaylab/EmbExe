### Estimation of metacell proliferation rates
Gives for each metacell a coarse estimate of its proliferation rate based on expression of cell-cycle-related genes.

```r
setwd(here::here())
```

```r
library("metacell")
```

```r
scdb_init("scrna_db/")
scfigs_init('figs')
```

```r
mat = scdb_mat("emb")
mc = scdb_mc("emb")
```

```r
source("scripts/embexe_cellcycle_analysis.r")
```

```r
cmp_cc = embexe_cc_plots(mat_id = "emb",mc_id = "emb",mc2d_id = "emb_umap",m_0 = 0.0075,s_0 = 0.0042)
```

```r
mc_fr = table(factor(x = mc@mc[cmp_cc],levels = c(1:ncol(mc@e_gc))))/as.vector(table(mc@mc))
```

```r
options(repr.plot.width = 36,repr.plot.height = 14)
barplot(mc_fr,col = mc@colors,border = T)
# rate is given in division per day 
# for a rate k, the expansion rate would be n = n0*2^(kt)
# default rate estimate is taken from Fig. 2B estimate in Mittnenzweig, Mayshar et al. of global cell division rate
default_proliferation_rate = 3.5

mc_rate = data.frame(mc = c(1:length(mc_fr)), proliferation_rate = default_proliferation_rate*as.vector(1 - mc_fr))

write.table(x = mc_rate,file = 'data/emb.metacell_proliferation_rate.tsv',sep = '\t',row.names = F)
```
