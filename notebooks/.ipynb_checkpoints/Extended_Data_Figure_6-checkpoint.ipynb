```r
library("metacell")
library("Matrix")
library(ggplot2)
library(ggrepel)
library("tgstat")
library("zoo")
```

```r
scdb_init("scrna_db/")
```

```r
source("scripts/atlas_proj.r")
```

```r
exutero_cell_type_annotation = read.table('data/exutero_cell_type_annotation.tsv',sep = '\t')
head(exutero_cell_type_annotation)
```

```r
mc_exutero = scdb_mc("exutero_f")
mc_wt = scdb_mc("embexe_recolored")
mc_wt@color_key
```

```r
embexe_to_emb = read.table("data/embexe_to_emb.old_to_new_metacell_ids.tsv",sep ='\t')
embexe_to_exe = read.table("data/embexe_to_exe.old_to_new_metacell_ids.tsv",sep ='\t')
embexe_to_emb_mc = embexe_to_emb$metacell_id_emb
names(embexe_to_emb_mc) = embexe_to_emb$metacell_id_embexe
embexe_to_exe_mc = embexe_to_exe$metacell_id_exe
names(embexe_to_exe_mc) = embexe_to_exe$metacell_id_embexe
```

```r
table(exutero_cell_type_annotation$experiment)
```

```r
exutero_hf_embryo_projected_metacells = exutero_cell_type_annotation$projected_metacell[ exutero_cell_type_annotation$experiment == "Head_fold" & exutero_cell_type_annotation$projected_color %in% mc_wt@color_key$color[1:41]] 
exutero_hf_exe_projected_metacells = exutero_cell_type_annotation$projected_metacell[ exutero_cell_type_annotation$experiment == "Head_fold" & exutero_cell_type_annotation$projected_color %in% mc_wt@color_key$color[43:50]] 

exutero_ls_embryo_projected_metacells = exutero_cell_type_annotation$projected_metacell[ exutero_cell_type_annotation$experiment == "Late_streak" & exutero_cell_type_annotation$projected_color %in% mc_wt@color_key$color[1:41]] 
exutero_ls_exe_projected_metacells = exutero_cell_type_annotation$projected_metacell[ exutero_cell_type_annotation$experiment == "Late_streak" & exutero_cell_type_annotation$projected_color %in% mc_wt@color_key$color[43:50]] 

exutero_hf_embryo_projected_metacells = embexe_to_emb_mc[as.character(exutero_hf_embryo_projected_metacells )]
exutero_hf_exe_projected_metacells = embexe_to_exe_mc[as.character(exutero_hf_exe_projected_metacells )]

exutero_ls_embryo_projected_metacells = embexe_to_emb_mc[as.character(exutero_ls_embryo_projected_metacells )]
exutero_ls_exe_projected_metacells = embexe_to_exe_mc[as.character(exutero_ls_exe_projected_metacells )]
```

```r
plot_projected_cells_on_mc2d_background(mc2d_id = "emb_umap",best_reference_metacell = exutero_hf_embryo_projected_metacells,cex.points = 1,main_tag = "Ex utero LHF embryonic cells")
plot_projected_cells_on_mc2d_background(mc2d_id = "exe_umap",best_reference_metacell = exutero_hf_exe_projected_metacells,cex.points = 1,main_tag = "Ex utero LHF extraembryonic cells")

plot_projected_cells_on_mc2d_background(mc2d_id = "emb_umap",best_reference_metacell = exutero_ls_embryo_projected_metacells,cex.points = 1,main_tag = "Ex utero LS embryonic cells")
plot_projected_cells_on_mc2d_background(mc2d_id = "exe_umap",best_reference_metacell = exutero_ls_exe_projected_metacells,cex.points = 1,main_tag = "Ex utero LS extraembryonic cells")
```

#### Next follows 2d projection of matched WT cells

```r
matched_wt_cells = read.table('data/exutero_matched_wt_cells.tsv')
head(matched_wt_cells)

mc_wt_emb = scdb_mc("emb")
mc_wt_exe = scdb_mc("exe")
table(matched_wt_cells$exutero_batch)
```

```r
wt_hf_cells = matched_wt_cells$cell[ matched_wt_cells$exutero_batch == "HF"]
wt_ls_cells = matched_wt_cells$cell[ matched_wt_cells$exutero_batch == "LS"]

wt_hf_cells_emb = intersect(wt_hf_cells,names(mc_wt_emb@mc))
wt_hf_cells_exe = intersect(wt_hf_cells,names(mc_wt_exe@mc))
wt_ls_cells_emb = intersect(wt_ls_cells,names(mc_wt_emb@mc))
wt_ls_cells_exe = intersect(wt_ls_cells,names(mc_wt_exe@mc))
```

```r
length(wt_hf_cells_exe)
```

```r
plot_projected_cells_on_mc2d_background(mc2d_id = "emb_umap",best_reference_metacell = mc_wt_emb@mc[wt_hf_cells_emb],cex.points = 1,main_tag = "Matched WT LHF embryonic cells")
plot_projected_cells_on_mc2d_background(mc2d_id = "emb_umap",best_reference_metacell = mc_wt_emb@mc[wt_ls_cells_emb],cex.points = 1,main_tag = "Matched WT LS embryonic cells")

plot_projected_cells_on_mc2d_background(mc2d_id = "exe_umap",best_reference_metacell = mc_wt_exe@mc[wt_hf_cells_exe],cex.points = 1,main_tag = "Matched WT LHF EXE cells")
plot_projected_cells_on_mc2d_background(mc2d_id = "exe_umap",best_reference_metacell = mc_wt_exe@mc[wt_ls_cells_exe],cex.points = 1,main_tag = "Matched WT LS EXE cells")
```
