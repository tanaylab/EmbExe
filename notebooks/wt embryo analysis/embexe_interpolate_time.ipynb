### Translate transcriptional ranks into developmental time

Based on the nominal harvest time of each embryo and its calculated transcriptional rank, we infer for each embryo a corresponding developmental time (ranging from ~E5.5 to E8.5). Developmental time is calculated through spline interpolation of nominal harvest time vs transcriptional rank

```r
library("metacell")
scdb_init("scrna_db/",force_reinit = T)
scfigs_init('figs')
```

```r
library(dplyr)
mat = scdb_mat("embexe")
mc = scdb_mc("embexe_recolored")
```

```r
# read new transcriptional rank of embexe embryos
embexe_nominal_time = read.csv('data/embexe.morphology_nominal_time.csv',stringsAsFactors = F)
embexe_rank = read.table('data/embexe.transcriptional_ranks.tsv',sep = '\t',h = T,stringsAsFactors = F)
# wt10_age = read.table('data/wt10_transcriptional_rank_developmental_time.txt',sep = '\t',h = T,stringsAsFactors = F)
# wt10_embryo = read.table('data/wt10.size_morphology_measurements.txt',sep = '\t',h = T,stringsAsFactors = F)
# wt10_age_all = left_join(wt10_age,wt10_embryo,by = "transcriptional_rank")
head(embexe_nominal_time)
head(embexe_rank)
```

```r
embexe_time = left_join(embexe_rank,embexe_nominal_time,by= "embryo")
```

```r
f = !is.na(embexe_time$embryo_nominal_time) & !is.na(embexe_time$transcriptional_rank)


time_interpolation_spline = smooth.spline(x = embexe_time$transcriptional_rank[f],
                                          y = embexe_time$embryo_nominal_time[f],spar = 0.85)

plot(embexe_time$transcriptional_rank[f],embexe_time$embryo_nominal_time[f],pch = 19,
     xlab= "Transcriptional Rank",
     ylab = "Embryo Nominal Time")

lines(time_interpolation_spline$x,time_interpolation_spline$y)
```

```r
embexe_time$developmental_time = NA
embexe_time$developmental_time[f] = time_interpolation_spline$y
```

```r
missing_ranks = embexe_time$transcriptional_rank[!f]

ranks_filtered = embexe_time$transcriptional_rank[f]

times_filtered = time_interpolation_spline$y
names(times_filtered) = ranks_filtered



for (r in missing_ranks) {
    
    next_lower_rank = max(ranks_filtered[ranks_filtered < r])
    next_higher_rank = min(ranks_filtered[ranks_filtered > r])
    
    next_lower_time = times_filtered[as.character(next_lower_rank)]
    next_higher_time = times_filtered[as.character(next_higher_rank)]
    
    time_inter = (r - next_lower_rank)/(next_higher_rank - next_lower_rank)*next_higher_time + 
                 (next_higher_rank - r)/(next_higher_rank - next_lower_rank)*next_lower_time
    
    f_r = embexe_time$transcriptional_rank == r
    
    
    embexe_time$developmental_time[f_r] = time_inter
    
    
    
}
```

#### Binning of ranked embryos into age groups

```r
new_age = embexe_time
rownames(new_age) = new_age$embryo
```

```r
mat = scdb_mat("embexe_with_e45")
mc = scdb_mc("embexe_with_e45")
```

```r
all_cells = names(mc@mc)[mc@colors[mc@mc] %in% mc@color_key$color[1:40] & !is.na(mat@cell_metadata[names(mc@mc),"transcriptional_rank"])]
ct_ag = table(mc@colors[mc@mc[all_cells]],mat@cell_metadata[all_cells,"transcriptional_rank"])
ct_ag_n = t(t(ct_ag)/colSums(ct_ag))
ct_ag_n = ct_ag_n[mc@color_key$color[1:40],]
options(repr.plot.width=30, repr.plot.height=6)

space_between = rep(0.3,ncol(ct_ag_n))
border_age_group = c(2.5,12.5,28.5,74.5,97.5,119.5,139.5,151.5,165.5,178.5,188.5,201.5,212.5,223.5,235.5,243.5)

space_between[as.integer(border_age_group + 0.5)] = 2

barplot(ct_ag_n,col = rownames(ct_ag_n),las = 2,space = space_between)
length(border_age_group)
```

```r
new_age_group = rep(1,nrow(new_age))
lower_border = 0.5
higher_border = 0.5
border_age_group = c(border_age_group,length(new_age_group) + 0.5)

for (n in 1:length(border_age_group)) {
    
    higher_border = border_age_group[n]
    
    ranks_f = c(1:nrow(new_age))[(c(1:nrow(new_age)) < higher_border) & (c(1:nrow(new_age)) > lower_border)]
    
    new_age_group[ranks_f] = n
    
    lower_border = border_age_group[n]
}
```

```r
embryos_f = embexe_time$embryo[!is.na(embexe_time$transcriptional_rank)]
embryos_ff = embexe_time$embryo[!is.na(embexe_time$transcriptional_rank) & !is.na(embexe_time$morphological_rank)]
options(repr.plot.width=10, repr.plot.height=10)


plot(x = new_age[embryos_ff,"transcriptional_rank"],y = new_age[embryos_ff,"morphological_rank"],
     col = new_age_group[new_age[embryos_ff,"transcriptional_rank"]],pch = 19,
     xlab = "Transcriptional rank",ylab = "Morphological rank")
for (i in 1:length(border_age_group)) {
    abline(v = border_age_group[i])
    if(i == 1) {
       text(y = 220,x = 0.5*(border_age_group[i] + 0.5),labels = i) 
    } else {
       text(y = 220,x = 0.5*(border_age_group[i] + border_age_group[i-1]),labels = i) 
    }
    
}

```

```r

cells_f = names(mc@mc)[(mc@colors[mc@mc] %in% mc@color_key$color[1:40]) & !is.na(mat@cell_metadata[names(mc@mc),"transcriptional_rank"])]

plot(as.vector(table(new_age_group[mat@cell_metadata[cells_f,"transcriptional_rank"]])),pch = 19, xlab = "Age group",
     ylab = "Number of cells")
table(new_age_group[mat@cell_metadata[cells_f,"transcriptional_rank"]])

plot(x = new_age[embryos_f,"transcriptional_rank"],y = new_age[embryos_f,"developmental_time"],col = new_age_group[new_age[embryos_f,"transcriptional_rank"]],pch = 19)
table(new_age_group)

```

```r
time_age_group = tapply(new_age$developmental_time,new_age_group,median)

options(repr.plot.width=6, repr.plot.height=6)
plot(as.vector(time_age_group),pch = 19,xlab = "Age group",ylab = "Developmental time")

```
